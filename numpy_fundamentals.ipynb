```python
# Note 
'''
| Method      | Copy or View  |
| ----------- | ------------- |
| `reshape()` | view (mostly) |
| `ravel()`   | view          |
| `flatten()` | copy          |

'''

'''
Copy vs View (simple meaning)

🟢 1. View → same data (shared memory)
A view does NOT create new data
It just gives another way to look at the same array
👉 If you change one → the other also changes

e,g
import numpy as np
arr = np.array([1, 2, 3])
view_arr = arr.reshape(-1)   # usually a view

view_arr[0] = 100

print(arr)       # 🔥 changed!
print(view_arr)
'''


'''
2. Copy → separate data
A copy creates new memory
Changes do NOT affect original array

e.g of copy -->
arr = np.array([1, 2, 3])
copy_arr = arr.flatten()   # copy

copy_arr[0] = 100

print(arr)       # ❌ unchanged
print(copy_arr)
'''
```
